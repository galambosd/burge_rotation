```python
import pyranges as pr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import time
# Display all the columns of dataframes
pd.set_option('display.max_columns', None)
```

# Load STR data

```python
within_300 = pr.PyRanges(pd.read_csv("data/within_300_ss_pyranges.csv"))
within_300.head()
```

```python
intron_internal = pr.PyRanges(pd.read_csv("data/intron_internal_pyranges.csv"))
intron_internal.head()
```

# Compare each STR pyranges to all the eCLIP peaks pyranges

```python
all_eCLIP_dict = {}

for filename in os.listdir("data/eclip/"):
    if filename.endswith(".bed"):
        dataset_name = filename.split(".")[0]
        current_pr = pr.read_bed("data/eclip/{}".format(filename))
        assert(dataset_name not in all_eCLIP_dict)
        all_eCLIP_dict[dataset_name] = current_pr
```

```python
start = time.time()
eclip_STR_internal = pr.count_overlaps(all_eCLIP_dict, intron_internal,
                                                   strandedness = False)
end = time.time()
print("Overlap count took {}s".format(end-start))
eclip_STR_internal.head()
```

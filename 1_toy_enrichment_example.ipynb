# Pick an RBP to use for a toy example
To do this, look at Supp. Data 4 in [Van Nostrand et al. (2020)](https://doi.org/10.1038/s41586-020-2077-3). 
Pick an RBP that...
1. Has many total IDR peaks
2. Has many peaks near 5' and 3' splice sites
3. If possible also has peaks in distal intron?

Choose KHSRP (K562 dataset has more peaks than HepG2)
ENCODE datasets ENCSR438GZQ (K562) and ENCSR366DGX (HepG2)

Open pyranges from `str_near_splice.ipynb`

```python
within_300 = pr.PyRanges(pd.read_csv("data/within_300_ss_pyranges.csv"))
within_300.head()
```

```python
intron_internal = pr.PyRanges(pd.read_csv("data/intron_internal_pyranges.csv"))
intron_internal.head()
```

Went to ENCODE website and got IDR peak bed file for KHSRP, K562 (ENCSR438GZQ), file ENCFF871BYW

```python
KHSRP_K562_idr = pr.read_bed("data/ENCFF871BYW.bed")
KHSRP_K562_idr.head()
```

```python
# let's find KHSRP peaks that overlap with STRs near splice sites
KHSRP_STR_ss = within_300.overlap(KHSRP_K562_idr, strandedness = False)
print(len(KHSRP_STR_ss))
KHSRP_STR_ss.head()
```

```python
# let's find KHSRP peaks that overlap with intron-internal STRS
KHSRP_STR_internal = intron_internal.overlap(KHSRP_K562_idr, strandedness = False)
print(len(KHSRP_STR_internal))
KHSRP_STR_internal.head()
```

```python
# compare the distribution of STR lengths overall and STR lengths that overlap with peaks
fig, ax = plt.subplots()
ax.hist(within_300.Length, histtype = "step", color = "xkcd:blue", 
        density = True, cumulative = True, label = "All STRs near ss")

ax.hist(KHSRP_STR_ss.Length, histtype = "step", color = "xkcd:red", 
        density = True, cumulative = True,
        label = "eCLIP peaks overlapping STRs near ss")
ax.legend()
```

# Download all the relevant eCLIP files

The following is a metadata file describing all eCLIP ENCODE files that are in cell lines and are in the narrowPeak bed format (these are peaks from each biological replicate and the IDR-called combined peaks). Of these, we want to select files that describe both replicates.

```python
eCLIP_meta = pd.read_csv("data/eclip_metadata.tsv", sep = '\t')
eCLIP_meta.head()
```

```python
file_url_list = eCLIP_meta[(eCLIP_meta['File assembly'] == "hg19") & (eCLIP_meta["Biological replicate(s)"] == "1, 2")]["File download URL"]
assert(len(file_url_list) == 223)
```

```python
file_url_list.to_csv("data/eclip_file_url.txt", sep = "\n", header = False, index = False)
```

```python
import pyranges as pr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import time
import json
# Display all the columns of dataframes
pd.set_option('display.max_columns', None)
```

# Load STR data

```python
within_300 = pr.PyRanges(pd.read_csv("data/temp/within_300_ss_pyranges.csv"))
within_300.head()
```

```python
intron_internal = pr.PyRanges(pd.read_csv("data/temp/intron_internal_pyranges.csv"))
intron_internal.head()
```

# Compare each STR pyranges to all the eCLIP peaks pyranges

```python
# prune some of the columns from our pyranges
columns = ['Chromosome', 'Start', 'End', 'class', 'Length', 'Strand', 'num_units', 'actual_repeat','gene_id',
           'gene_strand','annotation','promoter','dist_to_tss','Distance']

eclip_STR_internal_df = intron_internal.as_df()[columns]
eclip_STR_internal_df = eclip_STR_internal_df.rename(columns = {'Distance': 'dist_to_ss', 'annotation':'str_annotation', 'promoter':'str_in_promoter'})

eclip_STR_within_300_df = within_300.as_df()[columns]
eclip_STR_within_300_df = eclip_STR_within_300_df.rename(columns = {'Distance': 'dist_to_ss'})
```

```python
times = np.array([])
i = 0
# all_eCLIP_dict = {}
eclip_id2rbp = {}

for filename in os.listdir("data/eclip/"):
    start = time.time()
    if filename.endswith(".bed"):
        dataset_name = filename.split(".")[0]
        current_pr = pr.read_bed("data/eclip/{}".format(filename))
        rbp_name = current_pr.as_df().iloc[0,3].rstrip('_IDR')
        # all_eCLIP_dict[dataset_name] = current_pr
        overlap_count = intron_internal.count_overlaps(current_pr, 
                                                       overlap_col = dataset_name, 
                                                       strandedness = False)
        eclip_id2rbp[dataset_name] = rbp_name
        col_to_add = overlap_count.__getattr__(dataset_name)
        eclip_STR_internal_df[dataset_name] = np.array(col_to_add)

    end = time.time()
    times = np.append(times, end-start)
    i+=1
    if i % 20 == 0:
        mean_time = np.mean(times)
        print(mean_time)
        print("{0} done, {1}s remaining".format(i, mean_time*(len(os.listdir("data/eclip/"))-i)))
```

```python
num_eclip_peaks = eclip_STR_internal_df.iloc[:,14:].sum()
```

```python
for rbp in num_eclip_peaks.sort_values(ascending=False)[:10].index.values:
    print(eclip_id2rbp[rbp])
```

```python
eclip_STR_internal_df.to_csv("data/temp/eclip_STR_internal_df.csv", index = False)
```
